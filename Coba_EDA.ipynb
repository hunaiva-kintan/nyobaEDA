# **Noise / Outlier**

**Noise** dapat terjadi karena:
- salah pengukuran
- kesalahan input
- dll

**Outliers**
adalah data yang signifikan berbeda dengan data lain.

- Bukan Noise karena datanya **Valid**
- Diolah secara terpisah, bukan di exclude
- Jumlah sedikit: lakukan EDA
- Jumlah banyak: bisa bikin model sendiri

***Kecenderungan perbedaan harga rumah akibat tipe tempat parkir***

```python
#Category Plot

p= sns.catplot(x="Parking", y="House_Price", data=df_1)
```

Category plot tidak bekerja karena terdapat outlier/noise.

Bagaimana cara mengetahui itu noise/outlier? dapat ditanya dengan orang yang memahami data

Jika itu noise -> dapat di exclude dari data

```python
#How to decide Outlier/Noise

#Univariate outlier removal
#Asumsi distribusi data (mis. dist normal)

p = sns.distplot(df_1['House_Price'], kde=True, rug=True)
```

```python
#mis. data berdistribusi normal
#gunakan confidence interval 95%

dfnorm = np.abs(df_1.House_Price - df_1.House_Price.mean()) <= (2*df_1.House_Price.std())

print(dfnorm.shape)
dfnorm.head(5)
```

True: didalam interval
False: tidak dalam interval

```python
df_3 = df_1[dfnorm] #Data tanpa outliers
print(df_3.shape, df_1.shape)
```

```python
#Distribusi data tanpa Outliers

p = sns.distplot(df_3['House_Price'], kde=True, rug=True)
```

```python
#Category Plot Tanpa Outliers

p= sns.catplot(x="Parking", y="House_Price", data=df_3)
```

**Tentang Interpretasi**

Kurang tepat untuk mengatakan bahwa **tipe parkir tidak mempengaruhi harga rumah** karena masih ada variabel lain yang dapat mempengaruhi harga.

Kalo mau ditambahkan: jika diasumsikan luas bangunan dll adalah sama, maka tipe parkir tidak mempengaruhi harga rumah.

# **MISSING VALUE**

Missing value ditangani apabila model yang digunakan **tidak dapat menghandle missing value**.

```python
#Melihat missing Value

print(df_3.isnull().sum())
```

```python
#Heatmap Missing Value

sns.heatmap(df_3.isnull(), cbar=False)
plt.title('Heatmap Missing Value')
plt.show()
```

```python
#Persentase Missing Value

(df_3.isnull().sum()/len(df_3)).to_frame('Persentase Missing Value')
```

```python
#Cara Menangani Missing Value jika MV tidak banyak
#Drop Rows

Xmiss = df_1.dropna() #MV minimal 1 di salah satu kolom, barisnya dihapus
df_3.dropna(how='all') #Jika ada MV di semua kolom, barisnya dihapus
df_3.dropna(thresh=2) #Jika MV min di 2 kolom, maka barisnya dihapus
df_3.dropna(subset=['Dist_Hospital'])[:7] #MV minimal 1 di kolom Dist Hospital
#Tambahkan inplace=true if really sure
df_3.dropna(inplace=True)

```

```python
#Sudah tidak ada Missing Value

print(df_3.isnull().sum())
```
